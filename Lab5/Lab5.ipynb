Question 1- Calculate anatically the Fourier Transform of x(t)

Question 2- Calculate the fourier transform of x(t) using fft function then compare the result to theorical one (calculated anatically in the previous question)

```python
import numpy as np
import matplotlib.pyplot as plt

f0 = 20
dt = 0.001
t = np.arange(0, 1, dt)
x = np.cos(2*np.pi*f0*t)
fftx = np.fft.fft(x)
absx = np.abs(fftx)
frequency = np.fft.fftshift(np.fft.fftfreq(len(t), d=dt))
plt.plot(frequency, absx)
plt.xlabel('frequency')
plt.show()
```
