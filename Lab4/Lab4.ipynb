Let x the analog signal defined as: x(t) = [sinc((t-t0)/T)]^2

Question 1- Plot the signal x for t0 = 3, T=2 in the interval [-2,8] with range 0.1


```python
import numpy as np
import matplotlib.pyplot as plt

t0=3
T=2
t=np.arange(-2,8,0.1)

x=np.sinc((t-t0)/T)**2

plt.plot(t,x)
plt.xlabel('t')
plt.ylabel('x(t)')
plt.title('The signal x(t)')
plt.show()
```

Question 2- Calculate the Fourier Transform of x using fft function then plot the module of the result

```python
x_shifted=np.fft.fftshift(x)
fft_x_shifted=np.fft.fft(x_shifted)
fft_x=np.fft.fftshift(fft_x_shifted)

plt.plot(fft_x)

plt.xlabel('frequency')
plt.ylabel('fft(x)')
plt.title('The Fourier Transform of x')
plt.show()
```
