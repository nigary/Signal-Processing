In this lab, we propose to calculate the Fourier transform of s(t)= (sin(2πt))^2 = 1/2 * ( 1 - cos(4πt))


Question 1- Plot in the same figure, the following functions: sin(2πt), cos(2πt), -1/2 * cos(4πt) and (sin(2πt))^2 for t in [0,1]

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 1, 1000)
#functions:
x = np.sin(2 * np.pi * t)
y = np.cos(2 * np.pi * t)
z = -(1/2) * np.cos(4 * np.pi * t)
m = (np.sin(2 * np.pi * t))**2
#plot
plt.grid(True)
plt.plot(t, x, label='sin(2πt)')
plt.plot(t, y, label='cos(2πt)')
plt.plot(t, z, label='-(1/2) * cos(4πt)')
plt.plot(t, m, label='(sin(2πt))^2')
plt.xlabel('t, time')
plt.legend()
plt.show()
```
