Let x(t) = 1/2A for x in [-A,A[ and 0 elsewhere.

Question 1- Plot x

```python
import numpy as np
import matplotlib.pyplot as plt

A=4
arange = np.arange(-4, 4, 0.01)
x = np.zeros(len(arange))
x[200:400] = 1/(2*A)
plt.plot(arange, x)

plt.xlabel('t')
plt.ylabel('x(t)')
plt.title(' x(t) = 1/2A for x in [-A,A[ and 0 elsewhere')
plt.show()
```


Question 2-Calculate the Fourier Transform using fft function and plot it

```python
from scipy.fft import fft, fftfreq
xshifted = np.fft.fftshift(x)
fftxShift = np.fft.fft(xshifted)
fftx = np.fft.fftshift(fftxShift)
plt.plot(fftx)

plt.title('Fourier transform of x(t)')
plt.xlabel('Frequency (Hz)')
plt.ylabel('|fourier_x(t)|')

plt.show()
```

Question 4-
Plot the Dirac comb


```python
T=1
f=1000
t = np.arange(-10, 10, 1/f)
#dirac comb function
dirac_comb=np.zeros_like(t)
dirac_comb[::int(f*T)] = 1
plt.plot(t, dirac_comb)
plt.title('Dirac comb')

```

Question 5- Let Xp the result of convolution of x by Dirac comb of period T >= 2.A. Calculate the convolution and plot the result

```python
Xp = np.convolve(x, dirac_comb, 'same') / f
plt.figure(figsize=(8, 5))
plt.plot(t, Xp)
plt.title('Convolution')
plt.ylabel('Amplitude')
plt.show()
```

Question 6- Calculate and plot the Fourier Transform of the resulting convolution product

---



```python
Xpfft = fft(Xp)
freqXp = fftfreq(len(t), t[1] - t[0])
m = (freqXp >= -40) & (freqXp <= 40)
plt.stem(freqXp[m], np.abs(Xpfft[m]))
plt.figure(figsize=(8, 5))
plt.show()
```
