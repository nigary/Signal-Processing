We aim to represent the analog signal (w(t)=cos(2πft)) in python, with f=0,2 Hz.


```python
#import necessary libraries
import numpy as np
import matplotlib.pyplot as plot
import time


```


Question 1 - define the frequency f and a table (n) with abscissae on which the signal will be calculated: one point every 0.1 s between 0 and 10 s



```python
f=0.2
n=np.arange(0,10,0.1)
```

Question 2- Calculate w(t) = cos(2πft) at elements of n, you can use np.pi to retrieve the exact value of pi.

```python
w_t=np.cos(2*np.pi*f*n)
```

Question 3-a- Plot the signal w, you can use the functions figure and plot from matplotlib.pyplot library to respectively create a figure and plot the signal. The figure function can have figsize as argument inorder to define the size of the figure. In order to add a label to the figure, you can add an argument 'label="cos"' to plot function. The function legend of matplotlib.pyplot library add the legend on the plot.
Question 3-b- Add the legend of x, y labels and the title of the plot. You can use the xlabel and ylabel functions of matplotlib.pyplot library to define the x and y labels, title function adds the the title.
You can use the values of n values by using `$n$`.
If needed, add the following function for display 'plot.show()'


```python
plot.figure(figsize=(10, 5))
plot.plot(n, w_t, label="w(t) = cos(2πft)")

plot.xlabel("Time (s)")
plot.ylabel("Amplitude")
plot.title("Analog Signal: $w(t) = \cos(2\pi ft)$ with f = 0.2 Hz")
plot.show()
```

Question 4- Edit the code inorder to represent the signal w, within one second, after shift in time by π/3, and a multiplication by 2. The frequency of the new signal has to be set to 5. Choose the adaquate time step in order to have a smooth curve.

```python
shifted_time=np.arange(0,1,0.001)
f=5
w_t=2*np.cos(2*np.pi*f*shifted_time+np.pi/3)
plot.plot(shifted_time,w_t,label="w(t) = 2*cos(2πft)")
plot.show()
```

Question 5- represent the damped signal Capture d’écran 2023-10-03 à 17.13.30.png

A=2, a=0.5, f=2, step time=0,05.


```python
A=2
a=0.5
f=2
step_time=0.05
t = np.arange(0, 12, step_time)
y_t = A * np.exp(-a * t) * np.cos(2 * np.pi * f * t)
plot.figure(figsize=(10, 6))
plot.plot(t, y_t, label="f(t) = A*e^(-at) · cos (2 πft)")
plot.xlabel("Time (s)")
plot.ylabel("Amplitude")
plot.title("Damped signal")
plot.show()
```

Question 7- Define, within -3 and 3,  the discrete function z[n] = A rect(n) with time step egal to 0.1.
(You can use the function 'where' of the numpy library)

```python
n = np.arange(-3, 3, 0.1)
z_n = A * np.where(np.abs(n) <= 0.5, 1, 0)
plot.stem(n, z_n)
plot.show()
```

Question 8- Represent the function a(n) = 0.5 - t for t in [-1/2,1/2] and 0 elsewhere in the intervall [-3,3] with time step egal to 0.1.


```python
a_n = np.where(np.logical_and(n >= -0.5, n <= 0.5), 0.5 - n, 0)
plot.stem(n, a_n)
plot.show()
```

Question 9- Compute the convolution product of a(t) by z(t) in the intervall [-3,3] with time step egal to 0.1.

---


(You can use the function convolve of numpy library with mode="same")

```python
conv = np.convolve(a_n, z_n, "same")
plot.plot(n, conv)
plot.show()
```
